### Simulation

Start with $n$ possible interventions. If we call the set of all interventions in the 95th percentile $x$, how long (in teacher-years) will it take us until our estimate $\hat{x}$ contains 80% of the members of the true $x$?

Parameters:
Individual student variation in growth scores

$\sigma_\epsilon = 0.6$ (Kaine and Staiger + EEF)

Classwide random variation in growth scores

$\sigma_\theta = 0.2$ (Chetty + Kaine and Staiger + EEF)

Average class size: 20 students (bum-pluck)

I am going to take the average effect of all the programs to be the "opportunity cost" of imposing new programs on all schools. See section 1 [here](https://morningpigeon.wordpress.com/2017/06/16/a-paradox/)

Initial prior on all programs will be similar to weak prior used above: $\mu = 0$, $\sigma=0.5$

Add negligible noise as a hack to permit initial calculation of the top 20

```python
N = 10
# Normal distn bayes update
def update(mp, sp, r, ss):
    mn = (r*sp**2+mp*ss**2)/(ss**2+sp**2)
    sn2 = (ss**2*sp**2)/(ss**2+sp**2)
    #print(np.array([mp,r,mn]).T)
    return mn, np.sqrt(sn2)

# Standard deviation of the results of one class - see assumptions above
sig = np.sqrt(0.2**2 + 0.25/20)

# Sample programs effects
def samp_progs(N):
    return np.sort(skewnorm.rvs(2.5,-0.08,0.27,N))

# Prior effect distributions
def gen_dist(N):
    return np.array([[i, 0 + np.random.random()*0.001, 0.5] for i in range(N)])

# Run n trials using uniform sampling
def run_trials(dists,progs,n,s=sig,**kwargs):
    d = np.copy(dists)
    if n <= len(dists):
        samp = np.random.choice(len(dists),
                                n,
                                replace=False)
    else:
        samp = [i for j in range(n//len(dists)) for i in range(len(dists))]
    cts= np.zeros(len(progs))
    for m in samp:
        cts[m] += 1
    samp = list(set(samp))
    s = s/np.sqrt(cts[samp])
    r = np.random.normal(progs[samp],s)
    d[samp,1], d[samp,2] = update(dists[samp,1], dists[samp,2], r ,s)
    return d, np.average(progs[samp])

# Run n trials using kinda-thompson sampling

def run_trials_ts(dists,progs,n,v=0.5,u=0.0,s=sig):
    d = np.copy(dists)
    samp = np.argmax(norm.rvs(loc=dists[:,1],scale=(v*dists[:,2]+u),size=[n,len(dists)]),axis=1)
    cts= np.zeros(len(progs))
    for m in samp:
        cts[m] += 1
    samp = list(set(samp))
    s = s/np.sqrt(cts[samp])
    r = np.random.normal(progs[samp],s)
    d[samp,1], d[samp,2] = update(d[samp,1],d[samp,2],r,s)
    return d, np.average(progs[samp])

# N - number of programs to evaluate
# nteach - number of teachers available to evaluate things
# policy - uniform or thompson(-ish) sampling
# sf - initial 'superforecaster' pass?

def eval_policy(N=100,nteach=10,policy=run_trials_ts,sf=False,vbase=0.8):
    logs = np.zeros((15,3))
    for i in range(1000):
        d = gen_dist(N)
        progs = samp_progs(N)
        if sf:
            d, _ = run_trials(d,progs,len(d),s=0.5)
        logs[0] += np.array([np.average(np.abs(d[:,1]-progs)),0,progs[np.argmax(d[:,1])]-np.average(progs)])/1000
        for j in range(1,15):
            d, p = policy(d,progs,nteach,v=0.5*vbase**j)
            logs[j] += np.array([np.average(np.abs(d[:,1]-progs)),
                                 p-np.average(progs),
                                 progs[np.argmax(d[:,1])]-np.average(progs)])/1000

    return logs
```

```python
logs_rand10 = eval_policy(N=100,nteach=10,policy=run_trials)
logs_rand100 = eval_policy(N=100,nteach=100,policy=run_trials)
logs_rand1000 = eval_policy(N=100,nteach=1000,policy=run_trials)
```

```python
plt.figure(figsize=(10,10))
plt.plot(range(11),logs_rand10[:,2],label='10 teachers')
plt.plot(range(11),logs_rand100[:,2],label='100 teachers')
plt.plot(range(11),logs_rand1000[:,2],label='1000 teachers')
plt.legend()
plt.xlim(-1,10)
plt.title("100 Programs under Evaluation",size='xx-large')
plt.xlabel("Years of evaluation",size='xx-large')
plt.ylabel("True effect size of estimated best program",size='xx-large')
plt.savefig('num_teachers.png')
```

```python
logs_rand10 = eval_policy(N=10,nteach=10,policy=run_trials)
logs_ts10 = eval_policy(N=10,nteach=10,policy=run_trials_ts)
logs_ts10_sf = eval_policy(N=10,nteach=10,policy=run_trials_ts,sf=True)
```

```python
plt.figure(figsize=(10,10))
plt.plot(range(11),np.concatenate([logs_rand10[:5,1],[logs_rand10[6,2] for i in range(6)]]),label='uniform sampling then best choice')
plt.plot(range(11),logs_ts10[:,2],label='optimistic sampling')
plt.title('10 Programs under Evaluation, 10 Teachers evaluating',size='xx-large')
plt.xlabel("Years of evaluation",size='xx-large')
plt.ylabel("Average effect size of implemented programs",size='xx-large')
plt.legend()
plt.xlim(-1,10)
plt.savefig('eval_strat_10.png')
```
